```python
import time
  
def timeis(func):
    '''Decorator that reports the execution time.'''
  
    def wrap(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
          
        print(func.__name__, end-start)
        return result
    return wrap
  
@timeis
def countdown(n):
    '''Counts down'''
    while n > 0:
        n -= 1

        
countdown(500000)
countdown(100001)
```

```python
a = 'global'
class Fred:
    a = 'class' # class scope
    b = (a for i in range(10)) # function scope
    c = [a for i in range(10)] # function scope
    d = a # class scope
    e = lambda: a # function scope
    f = lambda a=a: a # default argument uses class scope

#     @staticmethod # or @classmethod, or regular instance method
    def g(): # function scope
        return a
print(Fred.a) # class
print(next(Fred.b)) # global
print(Fred.c[0]) # class in Python 2, global in Python 3
print(Fred.d) # class
print(Fred.e()) # global
print(Fred.f()) # class
print(Fred.g()) # global
```

```python
names=[['animesh',23],['manas',22],['jhon',12],['amy',13]]
def insert():
    inp_name=input("Name")
    inp_age=int(input("age"))
    
    names.append([inp_name,inp_age])
    
    
def Sort():
    l = len(names)
    for i in range(0, l):
        for j in range(0, l-i-1):
            if (names[j][1] > names[j + 1][1]):
                temp = names[j]
                names[j]= names[j + 1]
                names[j + 1]= temp
    print(names)

def searchByage(age):
    p=[]
    l = len(names)
    for i in range(0, l):
        if (names[i][1] == age):
            p.append(names[i])
    for _ in range(len(p)):
        print("Name",p[_][0])
        print("age",p[_][1])
    
def searchByname(name):
    p=[]
    l = len(names)
    for i in range(0, l):
        if (names[i][0] == name):
            p.append(names[i])
    for _ in range(len(p)):
        print("Name",p[_][0])
        print("age",p[_][1])
    
def printnames():
    print(names)
```

create a class rectagle, take two numerical arguements l and b, add metods to compute the area and perimeter of the rectangle
functions to return l and b. add a method issquare returns bool if rect is square

```python
class Rect:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
        
    def area(self):
        print("the area of the rectangle is: ", self.length*self.breadth)
        
    def perimeter(self):
        print("perimeter of the rectangle is: ", 2*(self.length+self.breadth))
    
    def is_square(self):
        if self.length==self.breadth:
            return True
        else:
            return False
    
```

```python
obj=Rect(12,13)
```

```python
obj.area()
```

```python
obj.perimeter()
```

```python
obj.is_square()
```

```python
def sum_range(A,B,M):
    s=0
    for i in range(A,B+1):
        if i%M:
            s+=i
    return s
```
